```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
ticker = "MSFT"
df = yf.download(ticker, start="2017-01-01")
```

```python
d1 = df["Close"].diff()
d2 = df["Close"].diff().diff()
df["d1"]=d1
df["d2"]=d2
```

```python
df = df.dropna().copy()
```

```python
df
```

```python
df = df.apply(lambda x: (x  / x.max() ))
```
